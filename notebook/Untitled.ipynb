```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
p = 50
q = 50
success_prob_s1 = 0.5
```

```python
W = np.random.normal(0, 1, (p, p))
K = np.random.binomial(n=1, p=success_prob_s1, size=(p, p))
```

```python
plot.matshow(W)
print('nonzero W-entry count: ', np.count_nonzero(W))
```

```python
plot.matshow(K)
print('nonzero K-entry count: ', np.count_nonzero(K))
```

```python
plot.matshow(W*K)
print('nonzero WK-entry count: ', np.count_nonzero(W*K))
```

```python
success_prob_s2 = 0.2
Q_idx = np.random.binomial(n=1, p=success_prob_s2, size=p)
Q = np.zeros((p, q))
Q[np.nonzero(Q_idx)[0],:] = 1
plot.matshow(Q)
print('nonzero K-entry count: ', np.count_nonzero(Q))
```

```python
B =  W * K * Q
plot.matshow(B)
print('nonzero B-entry count: ', np.count_nonzero(B))
```
